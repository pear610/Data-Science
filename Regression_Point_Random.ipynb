```
import csv
import os
import numpy as np

data_path = r"/content/drive/My Drive/GoogleColab/Regression/population.csv"

with open(data_path) as File:
  csv_reader = csv.reader(File, delimiter=',')
  
  data=[]
  eards=[]
  population=[]
  global_region="d"
  region="d"
  region_num = 0
  for row in csv_reader:
    if global_region == row[0] and region == row[1]:
      eards.append(int(row[2]))
      population.append(int(row[3]))
    else:
      if len(eards)>0:
        region_num+=1
        # data.append([eards, population])
        data.append([np.array(eards), np.array(population)])
        eards=[]
        population=[]
    global_region=row[0]  
    region=row[1]
  print(region_num)
  
graph_num = 40
# graph_num = len(data)
fig, axes = plt.subplots(nrows=graph_num, ncols=1, figsize=(10,60), dpi=300)

for i,ax in enumerate(axes):
  if i<=graph_num:
    ax.plot(data[i][0], data[i][1], 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Заголовок')
  else:
    break

fig    
plt.tight_layout()
```

```
# теперь берём один ряд и нарезаем его скользящим окном размером 20 элементов

import random
random.seed()
x = random.randint(0,len(data))
#print(data)
#print(data)
data1=data[x]
#print(data1)


input_len = 20
piece_num = (len(data1[1])- input_len - 5)
data_x_list = []
data_y_list = []
for i in range(piece_num):
  vekt_x = np.reshape(data1[1][i:i+input_len], (input_len, 1))
  data_x_list.append(vekt_x)
  data_y_list.append(data1[1][i+input_len:i+input_len+5])

data_array = np. stack(data_x_list)
data_y_array = np. stack(data_y_list)
# print(data_y_array)
# print(data_y_array.shape)
# print(data_array)
# print(data_array.shape)

print(data_array)

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, BatchNormalization, LeakyReLU
from tensorflow.keras.layers import Activation, Input, MaxPooling1D, Dropout
from tensorflow.keras.layers import AveragePooling1D, Conv1D, Flatten
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.optimizers import Adam, RMSprop, SGD

from tensorflow.keras.utils import plot_model
from IPython.display import display, Image

def my_plot(model):
  plot_model(
      model,
      to_file="model.png",
      show_shapes=True,
      show_layer_names=True,
      rankdir="TB",
      #expand_nested=False,
      #dpi=96
  )
  image = Image('model.png')
  display(image)

model = Sequential()
model.add( Conv1D( filters=32, kernel_size=5, padding="same", strides=1,
                   input_shape= (input_len, 1)))
#                  (batch_size, sequence_length, num_features)

# model.add( Conv1D( 32, 4, activation='relu',input_shape=input_shape[1:])
model.add( Conv1D( 8, 5))
model.add( Dropout(0.3))
model.add( Conv1D( 16, 5))
model.add( Dropout(0.3))
model.add(Flatten())
model.add(Dense(16, activation='relu'))
model.add(Dense(5, activation=None))

my_plot(model)

#optimizer = SGD(learning_rate=0.0001)
#optimizer = tf.keras.optimizers.RMSprop(0.001)
optimizer = Adam(learning_rate=0.0001, beta_1=0.9, beta_2=0.999, amsgrad=False)

model.compile( optimizer=optimizer ,
               loss = 'mae'
             )

EPOCHS = 1000

model.fit( data_array, data_y_array,
    epochs=EPOCHS,
    batch_size=2
)

print("Model trained")

# rand_data = np.random.randint(0, 3, (4, input_len, 1))
# pred = model.predict(rand_data)

# pred = model.predict(data_array)
# print (pred)


val_data = np.array(data1[1][len(data1[1])-input_len-5:len(data1[1])-5])
#print(len(data1[1]))
#print(val_data.shape)
val_data = np.reshape(val_data, (1, input_len, 1))
val_arg=data1[0][len(data1[0])-5:]
# print(val_data.shape)
# print(val_data)
pred = model.predict([val_data])

plt.plot(data1[0], data1[1], 'g', val_arg, pred[0], 'bo')
```
